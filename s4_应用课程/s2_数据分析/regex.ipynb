```python
# complete codes
import re
with open('./index.html','r') as fob:
    f=fob.read()
patternJs =re.compile('novel/[a-zA-Z-_]{1,20}.js', re.S)
retListJs=re.findall(patternJs, f)
for item in retListJs:
    print(item)
    new_item='{{% static "{}" %}}'.format(item)
    print(new_item)
    f=f.replace(item,new_item)# replaced must be arrange!
    
print('*'*20)
patternCss =re.compile('novel/[a-zA-Z-_]{1,20}.css', re.S)
retListCss=re.findall(patternCss, f)
for item in retListCss:
    print(item)
    new_item='{{% static "{}" %}}'.format(item)
    print(new_item)
    f=f.replace(item,new_item)
    
    
with open('./newindex.html', 'w')as fob2:
    fob2.write(f)
# print(f)
```
